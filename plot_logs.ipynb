```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path_extend = 'training_csv/run-'
a2c_ant = pd.read_csv(path_extend + 'A2C_0-tag-rollout_ep_rew_mean.csv')
a2c_quad = pd.read_csv(path_extend + 'A2C_quad-ex_0-tag-rollout_ep_rew_mean.csv')

ars_ant = pd.read_csv(path_extend + 'ARS_Ant-v5_0-tag-rollout_ep_rew_mean.csv')
ars_quad = pd.read_csv(path_extend + 'ARS_Quad-v5_0-tag-rollout_ep_rew_mean.csv')

ddpg_ant = pd.read_csv(path_extend + 'DDPG_0-tag-rollout_ep_rew_mean.csv')
ddpg_quad = pd.read_csv(path_extend + 'DDPG_quadex_0-tag-rollout_ep_rew_mean.csv')

ppo_ant = pd.read_csv(path_extend + 'PPO_0-tag-rollout_ep_rew_mean.csv')
ppo_quad = pd.read_csv(path_extend + 'PPO_quad-ex_0-tag-rollout_ep_rew_mean.csv')

sac_ant = pd.read_csv(path_extend + 'SAC_0-tag-rollout_ep_rew_mean.csv')
sac_quad = pd.read_csv(path_extend + 'SAC_quadex_0-tag-rollout_ep_rew_mean.csv')

all_csv = [a2c_ant, a2c_quad, ars_ant, ddpg_ant, ddpg_quad,
            ppo_ant, ppo_quad, sac_ant, sac_quad]
all_labels = ['A2C_Ant', 'A2C_Quad', 'ARS_Ant', 'DDPG_Ant',
              'DDPG_Quad', 'PPO_Ant', 'PPO_Quad', 'SAC_Ant', 'SAC_Quad']
all_data = zip(all_csv, all_labels)
```

```python
def plot_data(df, x_label, y_label, save_path):
    """
    This function takes a DataFrame and plots the data.
    :param df: DataFrame containing the data to plot
    :param x_label: Label for the x-axis
    :param y_label: Label for the y-axis
    """

    plt.figure(figsize=(10, 6))
    plt.plot(df[x_label], df[y_label], linestyle='-')
    title = save_path.split('_')
    plt.title(f'{title[0]} on {title[1]}')
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.grid()
    plt.savefig('training_plots/' + save_path + '.png')
```

```python
for (df, save_path) in all_data:
        plot_data(df, 'Step', 'Value', save_path=save_path)
```
